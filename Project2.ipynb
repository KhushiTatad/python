#  ***Gender***

```python
plt.figure(figsize=(10, 5))
gender = {'F': 'pink', 'M': 'blue'}
ax = sns.countplot(x='Gender', data=df, palette=gender, edgecolor = 'black', linewidth=2)
plt.title('Gender Distribution',fontsize='14',fontweight='bold')
plt.xlabel('Gender',color= 'black',fontsize='12',fontweight='bold')
plt.ylabel('Count' ,color ='black',fontsize='12',fontweight='bold' )
sns.set_style("whitegrid")

for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)


gender = {'F': 'pink', 'M': 'blue'}
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen,color = 'pink', edgecolor = 'black', linewidth=2)
sales_gen
```

 ***From Above Graph We Can See That Most Of Buyer Are Female And Even Purchasing Power Of Females Are Greater Than Man***

```python
gender_colors = {'F': 'hotpink', 'M': 'black'}
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', palette=gender_colors, edgecolor = 'black', linewidth=2)
# df.describe()
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age,  color = 'black')
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = True)

sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age,  color = 'black')
```

***From Above Graph We Can See That Most Of The Buyers Are Of Age Group Between 26 - 35 years Female.***
