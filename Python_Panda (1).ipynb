```
import pandas as pd
import random
import numpy as np


brands = ['Toyota', 'Honda','Toyota', 'BMW', 'Nissan',"Toyota","Honda","Honda","Toyota","Nissan"]
colors = ['White', 'Red', 'Blue','black','white', 'Green','Blue','Blue','White','White']


data = {
    'Make': random.choices(brands, k=10),
    'Colour': random.choices(colors, k=10),
    'Odometer': np.random.randint(1000, 10000, 10).tolist(),
    'Sale_Price': np.random.uniform(3500, 6000, 10).round(2)
}

df = pd.DataFrame(data)


print(df)
df.to_csv('car_sales.csv')
```
