```python
set1 = {1, 2, 3, 4}
set2 = set([3, 4, 5, 6])

print("Set1:", set1)
print("Set2:", set2)
```

```python
s = {1, 2, 3}

s.add(4)
print("After add:", s)

s.remove(2)
print("After remove:", s)

s.discard(10)  # no error if not present
print("After discard:", s)

s.pop()
print("After pop:", s)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print("Union:", a.union(b))
# OR
print("Union using |:", a | b)
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print("Intersection:", a.intersection(b))
# OR
print("Intersection using &:", a & b)
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print("A - B:", a.difference(b))
print("B - A:", b - a)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print("Symmetric Difference:", a.symmetric_difference(b))
# OR
print("Using ^:", a ^ b)
```

```python
a = {1, 2}
b = {1, 2, 3, 4}

print("A is subset of B:", a.issubset(b))
print("B is superset of A:", b.issuperset(a))
```

```python
s = {1, 2, 3}

print(2 in s)
print(5 in s)
```

```python
lst = [1, 2, 2, 3, 4, 4, 5]

unique = list(set(lst))

print("Original:", lst)
print("Without duplicates:", unique)
```

```python
squares = {x**2 for x in range(1, 6)}
print("Squares:", squares)


evens = {x for x in range(10) if x % 2 == 0}
print("Even numbers:", evens)
```

```python
s = {1, 2, 3}

s.clear()
print("After clear:", s)
```
