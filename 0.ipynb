```python
print("**********Area Calculator**********")

print("""press 1 to get the area of square
press 2 to get the area of rectangle
press 3 to get the area of triangle
press 4 to get the area of circle
 """)


choice = int(input("enter your choice: "))
if choice ==1:
  side = int(input("enter the side of square: "))
  area = side * side
  print("area of square is: ",area)
elif choice ==2:
  length = float(input("enter the length of rectangle: "))
  breadth = float(input("enter the breadth of rectangle: "))
  r_area = length * breadth
  print("area of rectangle is: ",r_area)

elif choice == 3:
  radius = float(input("enter the radius of triangle: "))
  height = float(input("enter the height of triangle: "))
  c_area = ((22/7)*(radius**2))
  print("area of circle is : ",c_area )

elif choice == 4:
  base= float(input("enter the base of triangle: "))
  height = float(input("enter the height of triangle: "))
  t_area = 0.5 *base*height
  print("area of circle is : ",t_area )
else:
  print("invalid choice")
```
