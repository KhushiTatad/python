```python
s = "Hello World"

print("String:", s)
print("First char:", s[0])
print("Last char:", s[-1])
print("Slice (0:5):", s[0:5])

for ch in s:
    print(ch, end=" ")
```

```python
a = "Hello"
b = "World"

print(a + " " + b)
print(a * 3)
```

```python
s = "hello world"

print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
```

```python
s = "hello world"

print(s.find("world"))
print(s.index("hello"))
print(s.startswith("he"))
print(s.endswith("ld"))
```

```python
s = "hello world"

new_s = s.replace("world", "Python")
print(new_s)
```

```python
name = "Jagdeep"
age = 21

# format()
print("My name is {} and I am {}".format(name, age))

# f-string
print(f"My name is {name} and I am {age}")

# % formatting
print("My name is %s and I am %d" % (name, age))
```

```python
s = "   hello   "

print(s.strip())    # remove spaces
print(s.lstrip())
print(s.rstrip())

# padding
print(s.strip().center(20, '*'))
print(s.strip().ljust(20, '-'))
print(s.strip().rjust(20, '-'))
```

```python
s = "apple,banana,grape"

lst = s.split(",")
print(lst)

joined = "-".join(lst)
print(joined)
```

```python
s = "banana"

print("Count of a:", s.count("a"))
print("Count of n:", s.count("n"))
```

```python
s1 = "abc"
s2 = "123"
s3 = "abc123"
s4 = "   "

print(s1.isalpha())
print(s2.isdigit())
print(s3.isalnum())
print(s4.isspace())
```
